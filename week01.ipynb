Transformation between two coordinate systems:

\end{align*}
In homogeneous coordinates:

\begin{align*}
    \begin{bmatrix}

We can show that the inverse transformation is:
\begin{align*}
    \textbf{R}' &= \textbf{R}^T\\

\end{align*}

## Lines
#### Equation of a line:
A line can be described by the equation:
\begin{align*}
    ax + by + c = 0
\end{align*}
We can write a line in homogeneous coordinates $\textbf{l} = \begin{bmatrix} a & b & c\end{bmatrix}^T$
Given a point in homogeneous coordinates, it belongs to the line if and only if:
\begin{align*}
    0 &= \begin{bmatrix}
        a\\b\\c
    \end{bmatrix} \cdot \begin{bmatrix}
    sx\\sy\\s
    \end{bmatrix}
    &= \textbf{l} \cdot \textbf{p}_h
\end{align*}

#### Distance to a line:
In general, the distance of a point $\textbf{p} = \begin{bmatrix} sx  & sy & s\end{bmatrix} ^ T$ to a line $\textbf{l} = \begin{bmatrix}
a & b & c
\end{bmatrix}^T$ is:
\begin{align*}
    \frac{\left|\textbf{l}^T \textbf{p}\right|}{s \sqrt{a^2 + b^2}}
\end{align*}

If $a^2 + b^2 = 1$ and $s = 1$, the distance is $\left|\textbf{l}^T \textbf{p}\right|$

The points on a line are the ones that have a distance of 0 to it.

#### Intersection between two lines:
\begin{align*}
    \textbf{q} = \textbf{l}_1 \times \textbf{l}_2
\end{align*}

#### Normal to the line: $\textbf{n} = \begin{bmatrix}
a\\b
\end{bmatrix}$

#### Projection:
Projection of a point $\textbf{q} = \begin{bmatrix}
x\\y\\1
\end{bmatrix}$ along the line $\textbf{l}$: $\textbf{n}^T \begin{bmatrix}
x\\y
\end{bmatrix}
$

The distance of a point X to a line is preserved by projecting the point on a line perpendicular to the first.

### 6. Camera matrix
# TODO explain camera matrices
$\textbf{A}$ has the correct format to be a camera matrix: given a 2D point $\textbf{p}$, the point $\textbf{q} = \textbf{A}\textbf{p}$ is the homogeneous point seen by the camera. We have:
\begin{align*}
	\textbf{q} = \textbf{Ap} = \begin{bmatrix}
	fx + s\delta_x\\
	fy + s\delta_y\\
	s
	\end{bmatrix}
\end{align*}
%z becomes the scale: the distance of an object does not change the resulting pixel given a direction. 
$f = 10$ is the focal length: the bigger it is, the more "zoomed in" the picture is, so it acts as a scale. $\delta_x$ and $\delta_y$ are translations.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def box3d(n):
    """Generate a set of points: 3D cross in cube."""
    coordinates = np.zeros([3, 15 * n])
    half_length = 0.5
    border = half_length * np.ones(n)
    variation = np.linspace(-half_length, half_length, n)
    elems = np.array([-border, border, variation])
    indices = np.array(
        [
            [0, 0, 2],
            [0, 1, 2],
            [1, 0, 2],
            [1, 1, 2],
            [0, 2, 0],
            [0, 2, 1],
            [1, 2, 0],
            [1, 2, 1],
            [2, 0, 0],
            [2, 0, 1],
            [2, 1, 0],
            [2, 1, 1],
        ]
    )
    for i in range(12):
        coordinates[:, i * n : (i + 1) * n,] = elems[indices[i]]
    for i in range(12, 15):
        coordinates[i % 3, i * n : (i + 1) * n] = variation
    return coordinates


def projectpoints(K, R, t, Q):
    """Return the projected points as a 2xn matrix."""
    _, n = Q.shape
    Q = np.vstack([Q, np.ones(n)])
    T = np.vstack([np.hstack([R, t])])
    P = K @ T @ Q
    P = P[0:2, :] / P[2, :]
    return P
```

```python
R = np.eye(3)
K = np.eye(3)
t = np.zeros([3, 1])
t[2] = 4


p2d = projectpoints(K, R, t, Q)

plt.figure()
plt.plot(p2d[0], p2d[1], ".");
```

### 13. Rotation

```python
theta = np.pi / 6
R = np.array(
    [[np.cos(theta), 0, np.sin(theta)], [0, 1, 0], [-np.sin(theta), 0, np.cos(theta)]]
)


p2d = projectpoints(K, R, t, Q)

plt.figure()
plt.plot(p2d[0], p2d[1], ".");
```
